```python
model_path = '/content/drive/MyDrive/AI_Society_Project/model.pkl'
scaler_path = '/content/drive/MyDrive/AI_Society_Project/scaler.pkl'
```

```python
import pandas as pd
import numpy as np
import joblib
```

```python
model = joblib.load(model_path)

# Optional scaler
try:
    scaler = joblib.load(scaler_path)
    use_scaler = True
except:
    use_scaler = False

print("Model loaded from Drive successfully!")
```

```python
columns = [
    "HighBP","HighChol","CholCheck","BMI","Smoker","Stroke","Diabetes",
    "PhysActivity","Fruits","Veggies","HvyAlcoholConsump",
    "AnyHealthcare","NoDocbcCost","GenHlth","MentHlth",
    "PhysHlth","DiffWalk","Sex","Age","Education","Income"
]
```

```python
def predict_custom(input_data, threshold=0.5):

    input_df = pd.DataFrame([input_data], columns=columns)

    if use_scaler:
        input_df = scaler.transform(input_df)

    prob = model.predict_proba(input_df)[0]

    prediction = 1 if prob[1] > threshold else 0
    confidence = max(prob)

    return prediction, confidence, prob
```

```python
sample_input = [
    1, 1, 1, 40, 1, 0, 0, 0, 0, 1,
    0, 1, 0, 5, 18, 15, 1, 0, 9, 4, 3
]

prediction, confidence, probs = predict_custom(sample_input, threshold=0.5)

print("Prediction:", prediction)
print("Confidence:", confidence)
print("Probabilities:", probs)
```

```python
display_score = "{:.2f}%".format(confidence * 100)
if confidence < 0.6:
    print("⚠️ Low confidence → Human review needed",display_score)

elif prediction == 1:
    print("⚠️ High risk → Doctor validation required",display_score)

else:
    print("✅ Low risk:",display_score)
```
